### ABC Analysis

### A Items (Top 70–80% of value)

#####  High value

#####  Low quantity

#####  Require tight control

#### B Items (Next ~15–25%)

#####  Medium value

#####  Need moderate control

#### C Items (Lowest ~5%)

#####  Low value

#####  Bulk items

#####  Very little control needed

```python
import pandas as pd

df = pd.read_csv("retail_inventory_dataset_v2.csv")

# Identify sales columns
sales_cols = [col for col in df.columns if "Sales_Week" in col]

# Total 12-week sales
df["Total_12_Week_Sales"] = df[sales_cols].sum(axis=1)

# Annualized sales (approx)
df["Annual_Sales"] = df["Total_12_Week_Sales"] * 4.33  # 52 weeks / 12

# Annual consumption value
df["Annual_Consumption_Value"] = df["Annual_Sales"] * df["Unit_Price"]

# Sort by consumption value
df = df.sort_values("Annual_Consumption_Value", ascending=False)

# Calculate cumulative percentage
df["Cumulative_Value"] = df["Annual_Consumption_Value"].cumsum()
df["Cumulative_Percent"] = 100 * df["Cumulative_Value"] / df["Annual_Consumption_Value"].sum()

# Assign ABC category
def assign_abc(percent):
    if percent <= 70:
        return "A"
    elif percent <= 90:
        return "B"
    else:
        return "C"

df["ABC_Category"] = df["Cumulative_Percent"].apply(assign_abc)

df[["SKU", "Product_Name", "Annual_Consumption_Value", "Cumulative_Percent", "ABC_Category"]].head(20)
```
